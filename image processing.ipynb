BITWISE AND OR XOR-SHOWS NON INTERSECTION PARTS, NOT

```python
import cv2 as cv
import numpy as np
image=cv.imread("C:\\Users\\ashly\\Downloads\\ginifab protractor 20231206201236.jpg")
blank=np.zeros((400,400),dtype="uint8")
rectangle=cv.rectangle(blank.copy(),(30,30),(370,370),255,-1)#b,l,colour,size of coverage
circle=cv.circle(blank.copy(),(200,200),200,255,-1)#center,r,same as above
cv.imshow('rectangle',rectangle)
cv.imshow('circle',circle)
#AND
bitwise_and=cv.bitwise_and(rectangle,circle)
#or
bitwise_or=cv.bitwise_or(rectangle,circle)
#XOR
bitwise_xor=cv.bitwise_xor(rectangle,circle)
#NOT 
bitwise_not=cv.bitwise_not(rectangle)
cv.imshow('bitwise_and',bitwise_and)
cv.imshow('bitwise_or',bitwise_or)
cv.imshow('bitwise_xor',bitwise_xor)
cv.imshow('bitwise_not',bitwise_not)
cv.waitKey(0)
```

COMPUTING HISTOGRAM GRAY SCALE

```python
import cv2 as cv
from matplotlib import pyplot as plt
image=cv.imread("D:\\AJRP\\1st trial\\0001347.png")#close to 256 is white max intensity
gray=cv.cvtColor(image,cv.COLOR_BGR2GRAY)
cv.imshow('gray',gray)
gray_hist=cv.calcHist([gray],[0],None,[256],[0,256])#list of images,channels,mask,no of bins
plt.figure()
plt.title("grayscalehistogram")
plt.xlabel('bins')
plt.ylabel('no of pixels')
plt.xlim([0,256])
plt.plot(gray_hist)
plt.show()
cv.waitKey(0)
```

COMPUTING HISTOGRAM FOR MASKED IMAGE GRAY SCALE

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
image=cv.imread("D:\\AJRP\\1st trial\\0001347.png")#close to 256 is white max intensity
gray=cv.cvtColor(image,cv.COLOR_BGR2GRAY)
cv.imshow('gray',gray)
blank=np.zeros(image.shape[:2],dtype='uint8')
circle=cv.circle(blank.copy(),(image.shape[1]//2+45,image.shape[0]//2),200,255,-1)
masked=cv.bitwise_and(gray,gray,mask=circle)
cv.imshow('masked',masked)
gray_hist=cv.calcHist([masked],[0],None,[256],[0,256])#list of images,channels,mask,no of bins
plt.figure()
plt.title("grayscalehistogram")
plt.xlabel('bins')
plt.ylabel('no of pixels')
plt.xlim([0,256])
plt.ylim([0,60000])
plt.plot(gray_hist)
plt.show()
cv.waitKey(0)
```

HISTOGRAM FOR COLORED IMAGE

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
image=cv.imread("D:\\AJRP\\1st trial\\0001347.png")
blank=np.zeros(image.shape[:2],dtype='uint8')
circle=cv.circle(blank.copy(),(image.shape[0]//2,image.shape[1]//2),200,255,-1)
masked=cv.bitwise_and(image,image,mask=circle)
cv.imshow('masked',masked)
color=('b','g','r')
plt.figure()
plt.title("grayscalehistogram")
plt.xlabel('bins')
plt.ylabel('no of pixels')
for i,col in enumerate(color):
    hist=cv.calcHist([masked],[i],None,[256],[0,256])
    plt.plot(hist,color=col)
    plt.xlim([0,256])
plt.show()
cv.waitKey(0)
```

LAPLACIAN,SOBELX,Y EDGE DETECTION


```python
import cv2 as cv
import numpy as np
img=cv.imread("D:\\AJRP\\1st trial\\0001347.png")
cv.imshow("image",img)
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)#changing to gray
lap=cv.Laplacian(gray,cv.CV_64F)
lap=np.uint8(np.absolute(lap))
cv.imshow('Gray',gray)
cv.imshow('Laplacian',lap)
#sobel
sobelx=cv.Sobel(gray,cv.CV_64F,1,0)
sobely=cv.Sobel(gray,cv.CV_64F,0,1)
combined_sobel=cv.bitwise_or(sobelx,sobely)
cv.imshow('Sobelx',sobelx)
cv.imshow('Sobely',sobely)
cv.imshow('combined_sobel',combined_sobel)
cv.waitKey(0)
```
